### Hypothetical Statement - 3 : **The correlation between Opening Prices and Closing Prices of Yes Bank stock is significantly stronger than the correlation between High Prices and Low Prices.**

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (\(H_0\)):** The correlation between Opening Prices and Closing Prices of Yes Bank stock is not significantly stronger than the correlation between High Prices and Low Prices.

**Alternative Hypothesis (\(H_1\)):** The correlation between Opening Prices and Closing Prices of Yes Bank stock is significantly stronger than the correlation between High Prices and Low Prices.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import numpy as np
import scipy.stats as stats

# Calculate correlation coefficients
r_OC = df_copy['Open'].corr(df_copy['Close'])
r_HL = df_copy['High'].corr(df_copy['Low'])

# Number of observations
n = len(df_copy)

# Apply Fisher's z-transformation
z_OC = np.arctanh(r_OC)
z_HL = np.arctanh(r_HL)

# Standard errors
SE_OC = 1 / np.sqrt(n - 3)
SE_HL = 1 / np.sqrt(n - 3)

# Z-test
z_diff = (z_OC - z_HL) / np.sqrt(SE_OC**2 + SE_HL**2)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_diff)))

# Print the results
print("Correlation between Opening and Closing Prices (r_OC):", r_OC)
print("Correlation between High and Low Prices (r_HL):", r_HL)
print("Z-test statistic (z_diff):", z_diff)
print("p-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The statistical test used to obtain the p-value in this code is the z-test. It assesses whether the difference between two correlation coefficients is statistically significant.

##### Why did you choose the specific statistical test?

The specific statistical test chosen here is the z-test for comparing two correlation coefficients. This test is appropriate because it assesses whether the difference between two correlation coefficients is statistically significant. In this case, we are comparing the correlation between opening and closing prices (r_OC) with the correlation between high and low prices (r_HL) to determine if one correlation is significantly stronger than the other.

### ML Model - 1 : **Lasso Regression**

Lasso Regression, short for Least Absolute Shrinkage and Selection Operator, is a type of linear regression that incorporates L1 regularization into the model. It is used for feature selection and regularization to prevent overfitting in high-dimensional datasets. In Lasso Regression, the objective function includes the sum of the absolute values of the coefficients multiplied by a regularization parameter (alpha) along with the residual sum of squares. This encourages sparsity in the coefficient values, effectively performing feature selection by driving some coefficients to zero. As a result, Lasso Regression produces sparse models where only the most important features are retained.

In summary:
- Lasso Regression helps prevent overfitting by penalizing large coefficient values through L1 regularization.
- It encourages sparsity in the coefficient values, effectively performing feature selection.
- The regularization parameter (alpha) controls the strength of regularization, with higher values leading to more regularization.
- Lasso Regression is particularly useful when dealing with high-dimensional datasets with many features, as it can automatically select the most relevant features and discard the rest.

### ML Model - 2 : **Linear Regression**

Linear regression is a fundamental statistical and machine learning technique used for modeling the relationship between a dependent variable and one or more independent variables. In a simple linear regression, there is only one independent variable, while in multiple linear regression, there are multiple independent variables.

The key components of linear regression include:

1. **Dependent Variable (Y)**: The variable that we want to predict or explain.
2. **Independent Variable(s) (X)**: The variable(s) used to predict the dependent variable.
3. **Coefficients (Weights)**: The coefficients represent the weights assigned to each independent variable. These coefficients are estimated during the model training process.
4. **Intercept (Bias)**: The intercept is the constant term in the linear equation. It represents the value of the dependent variable when all independent variables are zero.
5. **Assumptions**: Linear regression assumes that there is a linear relationship between the independent and dependent variables, the errors are normally distributed, and there is homoscedasticity (constant variance of errors).
6. **Evaluation Metrics**: Common evaluation metrics for linear regression include mean squared error (MSE), root mean squared error (RMSE), mean absolute error (MAE), R-squared (R2) score, and adjusted R-squared score.

Linear regression is widely used in various fields such as economics, finance, biology, and social sciences for tasks such as prediction, forecasting, and inference. It provides a simple and interpretable way to understand the relationship between variables and make predictions based on observed data. However, it is important to ensure that the assumptions of linear regression are met before interpreting the results.

### ML Model - 3 : **Ridge Regression**

Ridge Regression is a linear regression technique used for modeling and predicting continuous variables. It is an extension of ordinary least squares regression that includes a penalty term called regularization. This penalty term helps to overcome issues such as multicollinearity and overfitting by adding a constraint to the coefficient values.

In Ridge Regression, the coefficients are estimated by minimizing the residual sum of squares (RSS) plus a penalty term, which is the L2 norm (sum of squares of coefficients). The penalty term is controlled by a hyperparameter called lambda (λ), also known as the regularization parameter. Increasing the value of lambda increases the amount of regularization applied to the model, which shrinks the coefficient estimates towards zero.
